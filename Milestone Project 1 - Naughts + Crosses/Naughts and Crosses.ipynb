```python
from IPython.display import clear_output
from time import sleep

#Function definition block

def select_symbol():
    #Lets players select their symbol
    
    player1 = None
    
    #Gets symbol for player 1
    while player1 not in ('X','O'):
        player1 = input('Player 1 select X or O: ').upper()
        
        #Checks symbol is X or O
        if player1 not in ('X','O'):
            print('Sorry please select only X or O.\n')
    
    #Sets player 2 to the other symbol
    if player1 == 'X':
        print('player 1 is X \nplayer 2 is O\n')
        return ('X','O')
    elif player1 == 'O':
        print('player 1 is O \nplayer 2 is X\n')
        return ('O','X')

def display_board(pos=[1,2,3,4,5,6,7,8,9]):
    #Function to print the board
    
    #Clears output every go
    clear_output(wait=True)
    
    #Prints the board
    print(f'\n{pos[0]} | {pos[1]} | {pos[2]}')
    print('----------')
    print(f'{pos[3]} | {pos[4]} | {pos[5]}')
    print('----------')
    print(f'{pos[6]} | {pos[7]} | {pos[8]}')
    
def get_move(board):
    #Gets players moves
    
    pos = None
    
    #Checks input is a legal move
    while pos not in range(10) or 0:
        
        pos = input('Please select a position to play (1-9): ')
        
        #Checks input can be an int (checks for no accidental spaces)
        try:
            pos = int(pos)
        except:
            print('Not an number. Try again!')
        
        #Rety if move is not legal
        else:
            if pos not in range(10) or 0:
                print('Please select a value in range 1-9')

            if board[pos-1] != ' ':
                print('Can not go there')
                pos = None
    
    return pos

def update_board(board, position, player):
    #Board update
    
    board[position-1] = player
    
    return board

def check_win(playerMoves):
    #Checks for a win
    
    win = None
    wins = [[1,2,3],
            [4,5,6],
            [7,8,9],
            [1,4,7],
            [2,5,8],
            [3,6,9],
            [1,5,9],
            [3,5,7]]
    
    #Cycle through wins and see if players moves have the win set.
    for x in wins:
        
        check = all(y in playerMoves for y in x)
        
        #If they have the win set - end game. 
        if check:
            win = True 
            break
    
    #If the player has playerd 5 moves and no win it must be a draw.
    if len(playerMoves) == 5 and win == None:
        win = 'draw'
        
    return win

def initialise(board, players):
    #Resets variables, asks players for symbols and shows moves.
    
    #Wipe the screen
    clear_output(wait=False)
    #Reset variables to start state
    board = [' ']*9
    players = {'p1':[' ',[]],
               'p2':[' ',[]]}
    
    #Select symbols and show space numbers
    players['p1'][0], players['p2'][0] = select_symbol()
    sleep(1)
    display_board()
    
    return board, players

def play_game(board, players):
    #Main flow of one game
    
    result = [None, None]
    
    #While no draw and no win keep playing
    while result[0] != True and result[0] != 'draw': 
            for p in players.values(): 
                
                #Record current player in results 
                result[1] = p[0] 
                
                #Get player move and record it in player variable
                move = get_move(board)
                p[1].append(move) 
                
                #Update the board and display
                board = update_board(board, move, p[0]) 
                display_board(board) 
                
                #Check for win or draw
                result[0] = check_win(p[1]) 
                
                #If a win or a draw - end game
                if result[0]: 
                    break
                elif result[0] == 'draw': 
                    break
                    
    return result
                    
def outcome(result, playerWins, players): 
    #Record game win and display winner
    
    #Display draw and record nothing
    if result[0] == 'draw':
        print('\nDraw!')
    
    #Record win and display winner
    elif result[0] == True:
        
        #Record win
        if result[1] == players['p1'][0]:
            playerWins['p1'] += 1
        elif result[1] == players['p2'][0]:
            playerWins['p2'] += 1
        
        #Display winner
        print(f'\nWinner {result[1]}!')
        print(f"\nPlayer 1 wins: {playerWins['p1']}")
        print(f"\nPlayer 2 wins {playerWins['p2']}")

    return playerWins

def play_again():
    #Checks if players want another game
    
    run = None
    
    #Checks for legal input
    while run not in ('Y','N'):
            run = input('\nKeep playing? (Y or N): \n').upper() 
            
            if run not in ('Y','N'):
                print('Please pick Y or N only.\n')
                
    return run
```

```python
def Main():
    
    #Variable Decleration
    run = 'Y'
    playerWins = {'p1':0,'p2':0}
    players = None
    board = None
    result = None
    
    #Run until players say stop
    while run != 'N':
        #Initialise
        board, players = initialise(board, players)

        #Play Game
        result = play_game(board, players)  
        
        #Outcome 
        playerWins = outcome(result, playerWins, players)
        
        #Continue Game?
        run = play_again()
            
Main()
```
