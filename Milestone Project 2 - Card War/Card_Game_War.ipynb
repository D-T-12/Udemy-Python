## Design
- Card Class: Blueprint for a single card including suit, rank and value for comparison.
- Deck Class: For generating a 52 card deck and being able to shuffle and split.
- Player Class 
- Game Logic

```python
import random

#Global Variables
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card():
    
    def __init__(self,rank,suit):
        self.rank = rank
        self.suit = suit
        self.value = values[rank] #No need to pass value into function as values is a global variable
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck():
    
    def __init__(self):
        #Generating the deck on instantiation
        self.all_cards = []
        
        for suit in suits:
            for  rank in ranks:
                created_card = Card(rank,suit)
                self.all_cards.append(created_card)
    
    def shuffle(self):
        #Shuffle works in place
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        #Removes a card from the deck
        return self.all_cards.pop()
```

```python
#Check Deck is working
new_deck = Deck()
new_deck.shuffle()

for card in new_deck.all_cards:
    print(card)
    
print("\n")
print(len(new_deck.all_cards))
print("\n")

one_card = new_deck.deal_one()
print(one_card)
print("\n")

print(len(new_deck.all_cards))
```

```python
class Player():
    
    def __init__(self,name):
        
        self.name = name
        self.all_cards = []
    
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def add_cards(self,new_cards):
        if type(new_cards) == type([]):
            #Multiple Cards
            self.all_cards.extend(new_cards)
        else:
            #Single Cards
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return f"player {self.name} has {len(self.all_cards)} cards"
```

```python
# Game Setup
player1 = Player('one')
player2 = Player('two')

new_deck = Deck()
new_deck.shuffle()

for x in range(26):
    player1.add_cards(new_deck.deal_one())
    player2.add_cards(new_deck.deal_one())

game_on = True
round_num = 0
```

```python
while game_on:
    
    #Round
    round_num += 1
    print(f"Round {round_num}")
    
    if len(player1.all_cards) == 0:
        print('Player One, out of cards, Player Two Wins!')
        game_on = False
        break
        
    if len(player2.all_cards) == 0:
        print('Player Two, out of cards, Player One Wins!')
        game_on = False
        break
        
    #Start a new round
    player1_cards = []
    player1_cards.append(player1.remove_one())
    
    player2_cards = []
    player2_cards.append(player2.remove_one())
    
    at_war = True
    
    while at_war:
            
            if player1_cards[-1].value > player2_cards[-1].value:
                player1.add_cards(player1_cards)
                player1.add_cards(player2_cards)
                
                at_war = False
            
            elif player2_cards[-1].value > player1_cards[-1].value:
                player2.add_cards(player2_cards)
                player2.add_cards(player1_cards)
                
                at_war = False
            
            else:
                print('War!')
                
                if len(player1.all_cards) < 5:
                    print('Player 1 out of cards\nPlayer 2 wins!')
                    game_on = False
                    break
                    
                elif len(player2.all_cards) < 5:
                    print('Player 2 out of cards\nPlayer 1 wins!')
                    game_on = False
                    break
                    
                else:
                    for num in range(5):
                        player1_cards.append(player1.remove_one())
                        player2_cards.append(player2.remove_one())
```
